# Importando pacotes

```python
# Importando alguns pacotes comuns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Pacotes de Machine Learning
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import load_boston
```

```python
import warnings
warnings.filterwarnings('ignore')
```

# Explorando os dados

```python
boston_precos = load_boston()
```

```python
X = pd.DataFrame(boston_precos.data, columns=boston_precos.feature_names)
Y = boston_precos.target
```

```python
print(X.shape)
X.head()
```

```python
print(Y.shape)
```

```python
print(boston_precos.DESCR)
```

# Visualizando os dados

```python
plt.figure(figsize=(12,8))
plt.scatter(X['RM'], Y, c="b", alpha=0.6)
plt.title("Preco mediano e numero de quartos", fontsize=16)
plt.xlabel("numero medio de quartos no imovel", fontsize=16)
plt.ylabel("preco mediano dos imoveis", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X['CRIM'], Y, c="r", alpha=0.6)
plt.title("Preco mediano e numero de criminalidade", fontsize=16)
plt.xlabel("crime per capita", fontsize=16)
plt.ylabel("preco mediano dos imoveis", fontsize=16)
plt.show()
```

# Modelo Machine Learning

```python
model = RandomForestRegressor()
model.fit(X,Y)
X['preco_previsto'] = model.predict(X)
```

# Olhar os resultados

```python
X.head()
```

```python
plt.figure(figsize=(12,8))
plt.scatter(X['preco_previsto'], Y, c="g", alpha=0.6)
plt.title("Preco mediano previsto e preco mediano verdadeiro", fontsize=16)
plt.xlabel("preco mediano previsto", fontsize=16)
plt.ylabel("preco mediano dos imoveis", fontsize=16)
plt.show()
```
